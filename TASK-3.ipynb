# Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# loading data set

```python
data = pd.read_csv("C:\\Users\\kr866\\Downloads\\householdtask3.csv")
```

# Bar Chart
with the data of year and income

```python
plt.bar(data["year"],data["income"])

plt.title("Bar Chart")

plt.xlabel("Year")
plt.ylabel("Income")

plt.show()
```

# line chart
with year and income

```python
plt.plot(data["year"])
plt.plot(data["income"])

plt.title("Line Chart")

plt.xlabel("Year")
plt.ylabel("Income")

plt.show()
```

```python
plt.scatter(data["year"],data["income"])

plt.title("Scatter plot")

plt.xlabel("Year")
plt.ylabel("Income")

plt.show()

```
